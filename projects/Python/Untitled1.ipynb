```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
URL= "https://www.snapdeal.com/products/mens-basketball-shoes?sort=plrty#bcrumbLabelId:255"
```

```python
page= requests.get(URL)
data=page.content
print(data)
soup=BeautifulSoup(data,'html.parser')
soup.prettify()
```

```python
product_name=soup.select(".product-title")
product_name
```

```python
P_name=[]
for i in product_name:
    P_name.append(i.get_text())
P_name
```

```python
product_price=soup.select(".lfloat.product-price")
product_price
```

```python
P_price=[]
for i in product_price:
    P_price.append(i.get_text())
P_price
```

```python
pro_price= [re.sub(r"Rs.  ","",item)for item in P_price]
pro_price
```

```python
dis=soup.select(".product-discount")
dis
```

```python
Discount =[]
for i in dis:
    Discount.append(i.get_text())
Discount
```

```python
discount=[re.sub(r"\n+",'',item)for item in Discount]
discount1=[re.sub(r"[a-zA-Z ]?",'',item)for item in discount]
discount1
```

```python
ratings=soup.select(".clearfix.rating.av-rating")
ratings
```

```python
rt=[]
for i in ratings:
    rt.append(i.get_text())
rt
```

```python
ratings_1 = [re.sub(r"[\n()]", '', item) for item in rt]
ratings_1
```

```python
actualprice=soup.select(".lfloat.product-desc-price.strike")
actualprice
```

```python
A_price=[]
for i in actualprice:
    A_price.append(i.get_text())
A_price
```

```python
Actual_price = [re.sub(r"Rs. ", '', item) for item in A_price]
Actual_price
```

```python
Basketball_shoes=pd.DataFrame({
    "Product Name": P_name,
    "Actual Price":Actual_price,
    "Discount":discount1,
    "Final Price":pro_price,
    "Ratings":ratings_1
})
Basketball_shoes
```
